### Preparing dataset for Model Training

```python
df_cleaned.head()
```

```python
X = df_cleaned.drop('MEDV',axis=1)
y = df_cleaned.iloc[:,-1]
X.head()
```

```python
X.columns
```

```python
y.head()
```

### Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test  = train_test_split(X,y,test_size=0.35, random_state=29) 
```

```python
X_train
```

```python
y_train
```

```python
y_test
```

```python
y_train
```

### Standardizing the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

### Model Training

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
print(regression.coef_)
```

#### Prediction with test data

```python
reg_pred = regression.predict(X_test)
reg_pred
```

#### Prediction plot

```python
plt.scatter(y_test,reg_pred)
```

### Residual errors

```python
residuals = y_test - reg_pred
residuals
```

```python
sns.displot(residuals,kde=True)
```

### Scatter plot wrt predictions and residuals

```python
#uniform distribution
plt.scatter(reg_pred,residuals)
```

### Performance metrics

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))

print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

#### R2 and adjusted R2

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
print("R2 score: ",score)

# Adjusted R2 score:
AR2 = 1- (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
print("Adjusted R2 score: ",AR2)
```
